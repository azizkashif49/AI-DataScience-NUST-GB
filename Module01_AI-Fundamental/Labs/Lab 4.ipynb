Task 6: Write a function is_bouncy that accepts an integer as argument and returns True if the
number is bouncy and False otherwise. Use this function in your program to print on screen all
the 4 digit bouncy numbers along with the total count of the 4-digit bouncy numbers.

```python
def is_bouncy(number):
    increasing = decreasing = False
    num_str = str(number)

    for i in range(1, len(num_str)):
        if num_str[i] > num_str[i - 1]:
            increasing = True
        elif num_str[i] < num_str[i - 1]:
            decreasing = True

        if increasing and decreasing:
            return True

    return False

is_bouncy(163879)
```

Task 8: Write a function called binom that takes two integers n and k and returns the binomial
coefficient.

```python
def binom(n, k):
    error = "Invalid input"
    
    if n < k:
        return error
    
    nk = n - k
    factorial_of_nk = 1
    for i in range(1, nk+1):
        factorial_of_nk = factorial_of_nk * i
    
    factorial_of_n = 1
    for i in range(1, n+1):
        factorial_of_n = factorial_of_n * i
        
    factorial_of_k = 1
    for i in range(1, k+1):
        factorial_of_k = factorial_of_k * i
    
    binomial = (factorial_of_n)/((factorial_of_k)*(factorial_of_nk))
    return binomial

n = int(input("Enter n: "))
k = int(input("Enter k: "))

print(binom(n, k))
```

 Write a function int2binary that accepts an integer as argument and returns its
equivalent binary number using while loop and a list
Use this function in a program that prints on screen the binary equivalent of numbers
from 100 till 1000 in a tabular form.
[Hint: Use the division-by-2 algorithm]


```python
def int2binary(num):
    binary = []
    while num > 0:
        binary.append(num % 2)
        num = num // 2
    binary.reverse()
    binary_str = ''.join(str(bit) for bit in binary)
    return binary_str

print(int2binary(int(input("Enter a number to convert into binary: "))))


print("\nBinary equivalent of numbers from 100 till 1000 in a tabular form")
print("\nDecimal ---------- Binary")    

for i in range(100, 1001):
    print(i, "              ", int2binary(i))
```
