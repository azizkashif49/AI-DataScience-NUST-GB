```python
# Importin required libraries
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

np.random.seed(7)
X = np.random.rand(1000, 2)
```

```python
# Step 1: Defining the number of clusters (K)
K = 2
```

```python
centroids = X[:K]
```

```python
centroids
```

```python
# Creating a function for K-means clustering
def kmeans_clustering(X, K, centroids):
    # Step 3: Assigning data points to the nearest centroid
    #formula x-x1 y2 -y1
    labels = np.argmin(np.linalg.norm(X[:, np.newaxis] - centroids, axis=2), axis=1)
    while True:
        # Step 4: Updating cluster centroids
        new_centroids = np.array([X[labels == k].mean(axis=0) for k in range(K)])
        # Checking convergence
        print("Old centroids --> ")
        print(centroids)
        print("New centroids --> " )
        print(new_centroids)
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
        # Step 3: Assigning data points to the nearest centroid
        labels = np.argmin(np.linalg.norm(X[:, np.newaxis] - centroids, axis=2), axis=1)
        return labels, centroids
```

```python
# Running K-Means Clustering
labels, centroids = kmeans_clustering(X,K,centroids)
```

```python
print(labels)
```

```python
# Visualizing the results
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='red')
plt.title("K-means Clustering")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
